```python
import pandas as pd

# Load the Glass dataset
df = pd.read_excel(r'C:\Users\Dell\Downloads\glass.xlsx')

# Display the first few rows of the dataset
print(df.head())
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print("Missing values in each column:\n", missing_values)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Box plot to visualize outliers
plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.title('Box Plot of Glass Dataset Features')
plt.show()
```

```python
# Histograms for each feature
df.hist(figsize=(12, 10), bins=20)
plt.suptitle('Histograms of Glass Dataset Features')
plt.show()
```

```python
print(df.dtypes)
```

```python
# Convert relevant columns to numeric, using coercion to handle any non-numeric values
df = df.apply(pd.to_numeric, errors='coerce')

# Check again the data types
print(df.dtypes)
```

```python
# Keep only numeric columns
numeric_df = df.select_dtypes(include=['number'])

# Create histograms for numeric columns
numeric_df.hist(figsize=(12, 10), bins=20)
plt.suptitle('Histograms of Glass Dataset Features')
plt.show()
```

```python
#Data Preprocessing
# Impute missing values with the mean for numeric columns
df.fillna(df.mean(), inplace=True)
```

```python
# One-hot encoding for categorical variables
df = pd.get_dummies(df, drop_first=True)
```
